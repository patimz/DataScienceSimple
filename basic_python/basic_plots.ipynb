# Basic Plots
In this notebook we will review 4 basic plots:

* Scatter plots
* Line plots
* Histograms
* Pie charts

But first we need to import some functions:

````python
import matplotlib.pyplot as plt
import numpy as np
import random
````

# Scatter plots

````python
#Create the data
x_val = [x for x in range(0, 100)]
y_val = [x*3 + 120*random.random() for x in range(0, 100)]

print(x_val[:3])
print(y_val[:3])
````

````python
#plot the data
plt.figure(figsize=(10,4))
plt.scatter(x_val, y_val, color="g")
plt.xlabel("This is the x axis", fontsize=18)
plt.ylabel("This is the y axis", fontsize=18)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.show()
````

# Regular plot (Line plot)

Now we will see how to create a line plot, a plot whit a continuous line. As an example we will plot two lines:

$y = 75x + 1$

and

$y = x^2 +2x + 1$

See below:

````python
#Generate y values for the new 2 curves
y_val2 = [x*75 + 1 for x in range(0, 100)]
y_val3 = [x*x + x*2 + 1 for x in range(0, 100)]

print(y_val2[:3])
print(y_val3[:3])
````

````python
#Plot the data
plt.figure(figsize=(5,5))
plt.plot(x_val, y_val2, color="g", label="y=75x + 1")
plt.plot(x_val, y_val3, color="b", label="y=x^2 +2x + 1")
plt.xlabel("x", fontsize=20)
plt.ylabel("y", fontsize=20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlim(0, 100)
plt.ylim(0, 8000)
plt.legend()
plt.show()
````

# Histograms

Let’s say we have the height of 10,000 individuals. In the block of code below we construct a histogram showing the height distribution:

````python
#generate the data
mu = 1.75
sigma = 0.1
height_data = np.random.normal(mu, sigma, 10000)

height_data[:3]
````

````python
#plot the histogram
plt.figure(figsize=(8,4))
plt.hist(height_data, bins=30, color="g")
plt.xlabel("Height (m)", fontsize=18)
plt.ylabel("Frequency", fontsize=18)
plt.xlim(1.4, 2.1)
plt.ylim(0, 1200)
plt.xticks(fontsize=14)
plt.yticks([0, 400, 800, 1200], fontsize=14)
plt.vlines(1.75, 0, 1200, colors="r")
plt.show()
````

# Pie charts

````python
cars_sold = [20, 30, 15, 25, 10]
car_makes = ["Toyota", "Mazda", "Honda", "Nissan", "Ford"]

plt.pie(cars_sold, 
        labels=car_makes, 
        colors=["b", "r", "g", "c", "m"], 
        shadow=True, 
        radius=1.2,
        explode = [0, 0.1, 0, 0, 0.2],
        textprops={'fontsize': 16}
       )
plt.show()
````

# Scatter plot explained
Ok, we got a plot! Now let’s break down what is happening line by line:

* ``` plt.figure(figsize=(10,4))```: this line sets the wide – long ratio. Try playing with the numbers yourself until you get a wide-long ratio that you like.

* ```plt.scatter(x_val, y_val, color="g")```: Here we tell python to generate the scatter plot. In the x axis we have ```x_val``` and in the y axis we have ```y_val```. Then, with the command ```color="g"``` we are telling python to make the plots green. "g" for green, but you can make it any color you want, check this [page](https://matplotlib.org/2.0.2/api/colors_api.html) for more infornation about it.

* ```plt.xlabel("This is the x axis", fontsize=18)```: This line creates the label for the x axis and sets the font size.

* ```plt.ylabel("This is the y axis", fontsize=18)```: This line creates the label for the y axis and sets the font size.

* ```plt.xticks(fontsize=18)```: This line sets the font size of the numbers in the x axis.

* ```plt.yticks(fontsize=18)```: This line sets the font size of the numbers in the y axis.

* ```plt.show()```: This line makes the plot

# Line plots explained

* ``` plt.plot(x_val2, y_val2, color="g", label="y=75x + 1")```: This line plots the curve $y = 75x + 1$, the color is set to green ```color=”g”```, and we are creating a label for the curve ``` label="y=75x + 1"```. This label shows up in the upper left corner of the plot.

* ``` plt.plot(x_val3, y_val3, color="b", label="y=x^2 +2x + 1")```: This line plots the curve $y = x^2 + 2x + 1$, the color is set to blue ```color=”b”```, and we are creating a label for the curve ``` label="y=x^2 +2x + 1"```. This label shows up in the upper left corner of the plot.

* ```plt.xlabel("x", fontsize=14)```: This line of code set the label of the x axis to **x** with a font size of 14

* ```plt.ylabel("y", fontsize=14)```: This line of code set the label of the y axis to **y** with a font size of 14

* ```plt.xticks(fontsize=14)```: This line sets the font size of the numbers in the x axis to 14

* ```plt.yticks(fontsize=14)```: This line sets the font size of the numbers in the y axis to 14

* ```plt.legend()```: This line of code creates the legends for the two curves. The ones in the top left corner.

* ```plt.show()```: This line of code makes the plot.

# Histogram plot explained

* ```height_data = np.random.normal(1.75, 0.1, 10000)```: Here the height data is generated. It is a **normal distribution**, centered at 1.75, with a standard deviation of 0.1, and we are generating 10,000 data points.

* ```plt.hist(height_data, bins=30, color="g")```: Here we are telling python that we want a histogram plot of the ``` height_data```, using 30 bins, and we want the bins to be green (```color=”g”```)

* ```plt.xlabel("Height (cm)", fontsize=18)```: Here we are telling python that the x axis will be labeled ** Height (cm)** and the font size is 18.

* ```plt.ylabel("Frequency ", fontsize=18)```: Here we are telling python that the y axis will be labeled **Frequency** and the font size is 18.

* ```plt.xlim(1.4, 2.1)```: This line sets the range of the x axis between 1.4 and 2.1

* ```plt.ylim(0, 1200)```: This line sets the range of the y axis between 0 and 1200

* ```plt.xticks(fontsize=14)```: This lines sets the font size of the numbers in the x axis to 14.

* ```plt.yticks([0, 400, 800, 1200], fontsize=14)```: This lines is telling python to only show the 0, 400, 800, and 1200 ticks in the y axis. Then, the font size if this ticks is set to 14.

* ```plt.vlines(1.75, 0, 1200, colors="r")```: This lines generates the solid red vertical line located at 1.75. The first number (1.75) tells python the position of the vertical line over the x axis, the second number (0) tells python the initial (y) position of the line, the third number (1200) tells python the final position (y) of the line. Finally ``` colors="r"``` is telling python that the line should be red.

* ```plt.show()```: This lines makes the plot
